# 선형 SVM (Linear SVM)

```python
#데이터 불러오기
import seaborn as sns #seaborn을 불러오고 네임스페이스를 sns로 축약
iris = sns.load_dataset('iris')  #iris라는 변수명으로 Iris 데이터셋 download

X = iris.drop('species', axis=1) #'species'열을 drop하고 input X를 정의
y_ = iris['species']
```

```python
#y data를 범주형으로 변환
from sklearn.preprocessing import LabelEncoder
classle = LabelEncoder() #class label
y = classle.fit_transform(iris['species'].values) #species 열의 문자열을 범주형으로 변환
```

```python
#전체 데이터를 학습과 테스트 데이터셋으로 분리
from sklearn.model_selection import train_test_split

#X(입력변수)와 y(종속변수)로 이루어진 데이터를 학습과 테스트 데이터넷으로 각각 70%, 30%의 비율로 나눔
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3, random_state=1, stratify=y) 
```

```python
#Linear SVM
from sklearn.svm import SVC 
svm = SVC(kernel='linear',C=1.0,random_state=1) #Linear SVM classifier 객체 생성
svm.fit(X_train,y_train) #모델 학습
y_train_pred = svm.predict(X_train) #학습데이터 예측
y_test_pred = svm.predict(X_test)   #시험데이터 예측
```

```python
print(svm.support_vectors_) #support vectors
print(svm.n_support_) #number of support vectors for each class
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_train, y_train_pred)) #학습데이터의 정확도
print(metrics.accuracy_score(y_test, y_test_pred))   #시험데이터의 정확도
```

```python
from sklearn import metrics
metrics.confusion_matrix(y_test,y_test_pred)  #confusion_matrix
```

```python
print(metrics.classification_report(y_test,y_test_pred)) #분류 성능요약
```

# 비선형 SVM (Nonlinear SVM)

```python
#데이터 불러오기
import seaborn as sns #seaborn을 불러오고 네임스페이스를 sns로 축약
iris = sns.load_dataset('iris')  #iris라는 변수명으로 Iris 데이터셋 download

X = iris.drop('species', axis=1) #'species'열을 drop하고 input X를 정의

#y data를 범주형으로 변환
from sklearn.preprocessing import LabelEncoder
classle = LabelEncoder() #class label
y = classle.fit_transform(iris['species'].values) #species 열의 문자열을 범주형으로 변환

#전체 데이터를 학습과 테스트 데이터셋으로 분리
from sklearn.model_selection import train_test_split

#X(입력변수)와 y(종속변수)로 이루어진 데이터를 학습과 테스트 데이터넷으로 각각 70%, 30%의 비율로 나눔
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3, random_state=1, stratify=y) 
```

```python
#Nonlinear SVM
from sklearn.svm import SVC 
ksvm = SVC(kernel='rbf', C=1.0, gamma=0.2, random_state=1) #RBF kernerl 객체 생성
ksvm.fit(X_train,y_train) #모델 학습
y_train_pred = ksvm.predict(X_train) #학습데이터 예측
y_test_pred = ksvm.predict(X_test)   #시험데이터 예측
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_train, y_train_pred)) #학습데이터의 정확도
print(metrics.accuracy_score(y_test, y_test_pred))   #시험데이터의 정확도
```

```python
from sklearn import metrics
metrics.confusion_matrix(y_test,y_test_pred)  #confusion_matrix
```

```python
print(metrics.classification_report(y_test,y_test_pred)) #분류 성능요약
```
